```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("Races.csv")

Race_data= df["Race (%)"]

USBorn = df["U.S. Born"]

plt.title("2019 Race Population for US Born \n\n\n\n\n\n\n\n\n")

# print(Race_data)

# print(USBorn)



fig1 = plt.pie(USBorn, labels=Race_data, autopct='%1.1f%%',pctdistance = 1.1,labeldistance = 1.3, shadow=True, startangle=90, textprops={'fontsize': 14}, radius =2)

fig = plt.gcf()
centre_circle = plt.Circle((0,0),0.20,fc='white')
fig.gca().add_artist(centre_circle)
plt.tight_layout
plt.show()

# What does one race mean in the data set?



```
