```python
R2010 = pd.read_csv("./data_cleaned/race population in 2010.csv")
R2020 = pd.read_csv("./data_cleaned/race population in 2020.csv")
```

```python
column_name = ['Geographic Area Name', 'total population_2010', 'White_2010',
       'Black or African American_2010', 'American Indian and Alaska Native_2010',
       'Asian_2010', 'Native Hawaiian and Other Pacific Islander_2010',
       'Some Other Race_2010', 'total population_2020', 'White_2020',
       'Black or African American_2020', 'American Indian and Alaska Native_2020',
       'Asian_2020', 'Native Hawaiian and Other Pacific Islander_2020',
       'Some Other Race_2020']
```

```python
Rtotal = R2010.merge(R2020,on = 'Geographic Area Name')
Rtotal = Rtotal.set_axis(column_name,axis = 1)
```

```python
index = Rtotal['Geographic Area Name'][0:14]
df = Rtotal.iloc[0:14,[1,8]].set_index(index).sort_values('total population_2010')
df
```

```python
x = df.plot.bar(rot =0)
x.set_ylabel("Population(in million)")
plt.xticks(rotation=90)
```
