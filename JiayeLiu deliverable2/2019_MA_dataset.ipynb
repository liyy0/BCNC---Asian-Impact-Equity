```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

dataset = pd.read_csv('2019MA_group.csv')
```

```python
Asian_groups=dataset.iloc[46:53,:]

print(dataset.dtypes)
dataset['Massachusetts!!Estimate'] = dataset['Massachusetts!!Estimate'].str.replace(',', '')

dataset['Massachusetts!!Estimate']=pd.to_numeric(dataset['Massachusetts!!Estimate'])
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(Asian_groups['Label (Grouping)'], Asian_groups['Massachusetts!!Estimate'], color ='maroon',
        width = 0.4)
plt.xlabel("Asian groups")
plt.ylabel("Population in 2019 MA")
plt.title("Specific population of different Asian groups in MA")
plt.show()
```

```python
dataset_A = pd.read_csv('2019_MA_Asian_SexbyAge.csv')
dataset_A['Massachusetts!!Estimate'] = dataset_A['Massachusetts!!Estimate'].str.replace(',', '')
dataset_A['Massachusetts!!Estimate']=pd.to_numeric(dataset_A['Massachusetts!!Estimate'])
dataset_A_agegroup=dataset_A.iloc[2:16,:]
dataset_A_agegroup_female = dataset_A.iloc[17:,:]
#print(dataset_A_agegroup_female)
fig = plt.figure(figsize = (15, 8))
X_axis = np.arange(len(dataset_A_agegroup['Label (Grouping)']))
plt.bar(X_axis -0.2, dataset_A_agegroup['Massachusetts!!Estimate'], label = 'Asian Male',color ='c',
        width = 0.2)
plt.bar(X_axis + 0.2, dataset_A_agegroup_female['Massachusetts!!Estimate'], label = 'Asian Female',color ='pink',
        width = 0.2)
plt.xticks(X_axis, dataset_A_agegroup['Label (Grouping)'])
plt.xticks(rotation=-30)
plt.xlabel("Asian Age Groups in MA")
plt.ylabel("Population")
plt.legend()
plt.title("Population of Asian in 2019 MA")
```
