```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df =  pd.read_csv("Median Household Income.csv")
df2 = pd.read_excel("co-est2019-annres-25.xlsx")
MAIncome = []
BostonIncome = []
USIncome = []
year = ['2013', '2014', '2015', '2016', '2017', '2018', '2019']
for i in range(len(df)):
    if(df['Geography'][i] == 'Massachusetts'):
        MAIncome += [df['Household Income by Race'][i]]
    if(df['Geography'][i] == 'Boston, MA'):
        BostonIncome += [df['Household Income by Race'][i]]
    if(df['Geography'][i] == 'United States'):
        USIncome += [df['Household Income by Race'][i]]
MAIncome.reverse()
BostonIncome.reverse()
USIncome.reverse()

plt.title('MA Income trend from 2013 to 2019')

plt.plot(year, MAIncome, label = "MAIncome")
plt.plot(year, BostonIncome, label = "BostonIncome")
plt.plot(year, USIncome, label = "USIncome")
plt.xlabel('Year')
plt.ylabel('Income')
plt.legend()
plt.show()






#MA Asian Population change from 2013 to 2019
df2 = df2.iloc[1, 4:11].tolist()
df2 = [388317,413079,430739,440845,452717,472403,472895]
fig, ax = plt.subplots()
plt.title('MA Population and Income Trend')
ax2 = ax.twinx()

ax.plot(year, df2, color = 'g', marker='o', label = 'MA Asian Population Change Over Time')
ax2.plot(year, MAIncome, color = 'b', marker='o', label = 'MA Income Change Over Time')
ax.legend()
ax2.legend(loc = 'lower right')
# giving labels to the axises
ax.set_xlabel('Year', color = 'r')
ax.set_ylabel('Population', color = 'g')

# secondary y-axis label
ax2.set_ylabel('Income', color = 'b')


 
# defining display layout
plt.tight_layout()

# show plot
plt.show()









df.head()
```
